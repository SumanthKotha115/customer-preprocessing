```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age': [25, np.nan, 35, np.nan, np.nan, 40],
    'city': ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})

# Remove duplicates
df = df.drop_duplicates()

# Fill missing values
df['age'].fillna(df['age'].median(), inplace=True)
df['city'].fillna(df['city'].mode()[0], inplace=True)

df
```

```python
from sklearn.preprocessing import LabelEncoder

# One-hot encode city
df = pd.get_dummies(df, columns=['city'])

# Label encode gender
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

df
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

cols = ['age', 'annual_income']

# Min-Max Scaling
df_minmax = df.copy()
minmax = MinMaxScaler()
df_minmax[cols] = minmax.fit_transform(df_minmax[cols])

print("Min-Max Scaled Data:")
print(df_minmax)

# Standard Scaling
df_standard = df.copy()
standard = StandardScaler()
df_standard[cols] = standard.fit_transform(df_standard[cols])

print("\nStandard Scaled Data:")
print(df_standard)
```

Min-Max Scaling is used when we want values between 0 and 1, useful for neural networks and distance-based models.

Standardisation is used when data follows a normal distribution and is better for algorithms like SVM and linear regression.
